## Problem 2. Plotting an Epicycloid
**2.A** $\color{magenta}{\text{Partner 3: Share your screen. You will do this portion of the problem with verbal help from partners 1 and 2. }}$

Define a function called `Epicycloid` which takes

* $r_1$, 
* $r_2$,
* $\theta$, 
* and vectorized function $H(r_1,r_2, \theta) = [f(r_1,r_2, \theta), g(r_1,r_2, \theta)]$

as inputs, where $r_1, r_2$ are scalars and $\theta$ is a vector, and returns a plot of $f$ as a function of $g$. 


```python
import matplotlib.pyplot as plt
```

```python
## code goes here 


def f(r1,r2,theta):
        return (r1+r2)*np.cos(theta) + r2* np.cos(theta * (1 + r1/r2) - np.pi)
def g(r1, r2, theta):
        return (r1+r2)*np.sin(theta) + r2* np.sin(theta * (1+ r1/r2) - np.pi)
    
def H(r1,r2,theta):
    return [f(r1,r2,theta),g(r1,r2,theta)]

def Epicycloid (r1,r2,theta, H):
    Hs = H(r1,r2,theta)
    fs = Hs[0]
    gs = Hs[1]
    return (plt.plot(gs,fs))

```

```python
r1 = 12
r2 = 7
theta = np.linspace(0,100*np.pi,1000)
Epicycloid (r1,r2,theta, H)
```
